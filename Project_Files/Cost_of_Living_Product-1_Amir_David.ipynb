```python
costofliving2 = pd.read_csv('./Project_Data/cost-of-living.csv')
```

```python
costofliving2 = costofliving2.transpose() #swap rows and columns
```

```python
costofliving2.columns = costofliving2.iloc[0] #assign row as column header
```

```python
costofliving2.reset_index(inplace=True) #convert index to column
```

```python
costofliving2.drop(0, axis=0, inplace=True) #delete row 0
```

```python
costofliving2.rename(columns = {'index':'City'}, inplace = True) #renaming column header
```

```python
costofliving2 = costofliving2.rename_axis(None, axis=1) #removing index header
```

```python
costofliving2
```

```python
#everything below is for costofliving1 in order!
```

```python
costofliving1 = pd.read_csv('./Project_Data/costofliving.csv')
```

```python
costofliving1 = costofliving1.drop(['Rank'],axis=1)
```

```python
costofliving1
```

```python
#Final merged data set below! 
```

```python
#merged set = 'merged_costofliving'
```

```python
merged_costofliving = pd.merge(costofliving1, costofliving2, on= 'City') #merging data sets
```

```python
merged_costofliving 
```

```python
#cleaning the data -- dropping missing and duplicate values for sanity
merged_costofliving.dropna(inplace = True) # dropping missing values
merged_costofliving.drop_duplicates(inplace = True) # dropping duplicate values here
merged_costofliving = merged_costofliving.reset_index(drop=True) # resetting the index

length = len(merged_costofliving.index) #save length of array of later
```

```python
merged_costofliving.shape
```

```python
merged_costofliving_mod1 = merged_costofliving.drop('Lettuce (1 head)', axis = 'columns')
```

```python
prep_df = merged_costofliving_mod1.drop("City", axis='columns')
merged_costofliving_mod2 = pd.concat([merged_costofliving_mod1, prep_df.astype(float)], axis = 'columns')
merged_costofliving_mod2.shape
```

```python
merged_costofliving_mod2 = merged_costofliving_mod1[~merged_costofliving_mod1['City']].astype(float)
```

```python
import seaborn as sns
```

```python
sns.heatmap(merged_costofliving.corr())
```

```python
X = merged_costofliving["Rent Index"].to_numpy()
y = merged_costofliving["Cost of Living Index"].to_numpy()
```

```python
X.shape[0]
```

```python
# Checking for Missing Data
if X.shape[0] != y.shape[0]:
  print("It looks like you have missing data. You may want to preprocess your data more with pandas to delete any rows with missing, NaN, N/A, and null values.")
  
idx = np.arange(length) #shuffle our dataset indices so we don't always split the same way!
np.random.shuffle(idx)

#split our data with 80% for training (learning) and 20% for testing.
split_threshold = int(length * 0.8)

train_idx = idx[:split_threshold]
# Uses the remaining indices for testing
test_idx = idx[split_threshold:]

# Generates train and test sets and formats them for training.
x_train, y_train = X[train_idx], y[train_idx]
x_test, y_test = X[test_idx], y[test_idx]
x_train= x_train.reshape(-1, 1)
y_train= y_train.reshape(-1, 1)
x_test = x_test.reshape(-1, 1)
```

```python
import matplotlib.pyplot as plt
```

```python
#let's plot our split data to see how it looks!
plt.figure(figsize=(10,5))

# plot the train set 
plt.subplot(1,2,1)
plt.scatter(x_train,y_train, c='orange')  
plt.xlabel('x', fontsize = 20) 
plt.ylabel('y', fontsize = 20)
plt.title('Generated Data - Train')
plt.grid('on')

# plot the test set 
plt.subplot(1,2,2)
plt.scatter(x_test, y_test)  
plt.xlabel('x', fontsize = 20) 
plt.ylabel('y', fontsize = 20)
plt.title('Generated Data - Test')
plt.grid('on')

plt.show()
```

```python
from sklearn.linear_model import LinearRegression
#Create the model object
linr = LinearRegression()
#Fit (train) the model -- this is where the ML happens!
linr.fit(x_train, y_train)
print(linr.intercept_, linr.coef_[0])
```

```python
# Predicting using SKLearn
y_hat = linr.predict(x_test)
```

```python
#plotting results
plt.figure(figsize=(10,5))
plt.plot(x_test, y_hat, '--')

plt.scatter(x_test,y_test, c='Green')  
plt.xlabel('x', fontsize = 20) 
plt.ylabel('y', fontsize = 20)
plt.title('Generated Data - Test')
plt.grid('on')
plt.show()
```

```python
print(linr.score(x_test, y_test))
```

```python
MAE = np.mean(abs(y_test - y_hat))
MAE
```
